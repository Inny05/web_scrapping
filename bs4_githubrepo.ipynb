```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html.parser 
import pprint
```

```python

url = "https://github.com/trending"
r= requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
htmlrepos = soup.find_all(class_="Box-row")

result = []
for repo in htmlrepos:
    author_repo = repo.find(class_= "h3").find("a").get_text().strip().replace('\n', '').replace(' ', '').split('/')
    nbr_stars = repo.find(class_= "f6").find('a').get_text().strip().replace('\n','').replace(',', '`').replace(' ','').split('/')
    result.append({'Developer': author_repo[0], 'Repository_name': author_repo[1], 'nbr_stars': nbr_stars[0]})

df=pd.DataFrame(result)
df.to_csv('Githubrepos.csv')  
```
